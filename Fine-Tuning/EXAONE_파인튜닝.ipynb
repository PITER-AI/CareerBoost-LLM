```python
import torch
import pandas as pd
import numpy as np
import warnings
import time
import chardet

import json
import jsonlines

from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    DataCollatorForLanguageModeling,
    BitsAndBytesConfig,
    TrainingArguments,
    pipeline,
    logging,
)

from peft import LoraConfig, get_peft_model
from trl import SFTTrainer
from huggingface_hub import notebook_login
from datasets import load_dataset
```

```python
# 모든 열을 Output 열에 삽입 후 기존 열 삭제
from datasets import load_dataset

# 파일 경로
file_path = "/content/gdrive/MyDrive/Colab Notebooks/PITer/datasets/smilestyle_dataset.tsv"

# 데이터셋 로드
dataset = load_dataset("csv", data_files=file_path, delimiter="\t")

# 데이터 전처리: 'Output' 열 생성 및 기존 열 삭제
def create_output_column(example):
    # 기존 열 데이터를 결합하여 'Output' 생성
    keys_to_merge = [
        'formal', 'informal', 'android', 'azae', 'chat', 'choding', 'emoticon',
        'enfp', 'gentle', 'halbae', 'halmae', 'joongding', 'king', 'naruto',
        'seonbi', 'sosim', 'translator'
    ]
    merged_text = " | ".join([f"{key}: {example[key]}" for key in keys_to_merge])
    return {"Output": merged_text}

# 'Output' 열 추가
dataset = dataset.map(create_output_column)

# 기존 열 삭제
keys_to_remove = [
    'formal', 'informal', 'android', 'azae', 'chat', 'choding', 'emoticon',
    'enfp', 'gentle', 'halbae', 'halmae', 'joongding', 'king', 'naruto',
    'seonbi', 'sosim', 'translator'
]
dataset = dataset.remove_columns(keys_to_remove)

# 결과 확인
print(dataset)
```

# **2. Load Pre-trained Language Model**

```python
peft_params = LoraConfig(
    lora_alpha=8, # 알파 / r -> scale
    lora_dropout=0.05, # 랜덤으로 5% 은닉함
    r=16,
    bias="none",
    target_modules=['q_proj', 'v_proj'],  # 모델 구조에 따라 조정
    task_type="CAUSAL_LM",
)
training_params = TrainingArguments(
    output_dir="./results",
    num_train_epochs=5,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    optim="paged_adamw_32bit",
    save_steps=25,
    logging_steps=25,
    learning_rate=1e-5,
    weight_decay=0.001,
    fp16=False,
    bf16=False,
    max_grad_norm=1.0,
    max_steps=-1,
    warmup_ratio=0.03,
    group_by_length=True,
    lr_scheduler_type="constant",
    report_to="none",  # W&B 비활성화
)
```

```python
from datasets import Dataset

# 모든 열 데이터를 병합하는 함수
def combine_fields(example):
    # 각 열 데이터를 "열 이름: 값" 형식으로 병합
    merged_text = " | ".join([f"{col}: {example[col]}" for col in example.keys()])
    return {"Output": merged_text}

# 데이터 전처리: 열 데이터를 결합한 "Output" 열 생성
dataset = dataset["train"].map(combine_fields)  # 'train' 스플릿 선택 후 전처리

# 결합된 데이터셋을 파일로 저장하는 함수
def save_preprocessed_dataset(dataset, output_file):
    # Dataset 객체를 Pandas DataFrame으로 변환
    df = dataset.to_pandas()

    # TSV 파일로 저장
    df.to_csv(output_file, sep="\t", index=False)
    print(f"Preprocessed dataset saved to {output_file}")

# 저장할 파일 경로 지정
output_file = "/content/gdrive/MyDrive/Colab Notebooks/PITer/datasets/clear_datasets.tsv"

# 결합된 데이터 저장
save_preprocessed_dataset(dataset, output_file)
```

```python
from transformers import Trainer, TrainingArguments
import torch

# SFTTrainer 초기화 및 학습
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=peft_params,
    dataset_text_field="Output",
    max_seq_length=None,
    tokenizer=tokenizer,
    args=training_params,
    packing=False,
)
trainer.train()

''' 이 코드는 캐시를 비우므로 바로 업로드 못함. 상황 보고 사용

# 학습 데이터 삭제 및 GPU 캐시 초기화
del dataset  # 학습 데이터셋 삭제
del trainer  # 트레이너 삭제
torch.cuda.empty_cache()  # GPU 캐시 초기화

print("Training data and trainer objects deleted, GPU cache cleared.")

'''
```
