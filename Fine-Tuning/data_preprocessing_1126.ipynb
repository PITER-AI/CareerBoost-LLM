```python
dataPath2 = "/content/data/traindataset.csv"

# Open the file in binary mode
with open(dataPath2, 'rb') as f:  # Changed dataPath to dataPath2
    # Detect the encoding
    result = chardet.detect(f.read())

# If chardet fails to detect the encoding, use 'latin1' as a fallback
encoding = result['encoding'] if result['encoding'] is not None else 'latin1'

dataset2 = pd.read_csv(dataPath2, encoding=encoding)
```

```python
def generate_prompt(example):
  output_texts = []
  for i in range(len(example['informal'])):
    if(str(example['informal'][i]) == "nan"):
      continue
    else:
      prompt = f"### Instruction: {example['informal'][i]} 이 문장을 정중한 말투로 바꿔줘.\n\n### Response: {example['formal'][i]}"
      prompt = prompt + "[|endofturn|]"
      output_texts.append(prompt)

  for i in range(len(example['chat'])):
    if(str(example['chat'][i]) == "nan"):
      continue
    else:
      prompt = f"### Instruction: {example['chat'][i]} 이 문장을 정중한 말투로 바꿔줘.\n\n### Response: {example['formal'][i]}"
      prompt = prompt + "[|endofturn|]"
      output_texts.append(prompt)

  for i in range(len(example['joongding'])):
    if(str(example['joongding'][i]) == "nan"):
      continue
    else:
      prompt = f"### Instruction: {example['joongding'][i]} 이 문장을 정중한 말투로 바꿔줘.\n\n### Response: {example['formal'][i]}"
      prompt = prompt + "[|endofturn|]"
      output_texts.append(prompt)

  for i in range(len(example['halmae'])):
    if(str(example['halmae'][i]) == "nan"):
      continue
    else:
      prompt = f"### Instruction: {example['halmae'][i]} 이 문장을 정중한 말투로 바꿔줘.\n\n### Response: {example['formal'][i]}"
      prompt = prompt + "[|endofturn|]"
      output_texts.append(prompt)

  return output_texts
```

```python
def generate_prompt2(example):
  output_texts = []
  for i in range(len(example['instruction'])):
    prompt = f"### Instruction: {example['instruction'][i]}\n\n### Response: {example['output'][i]}"
    prompt = prompt + "[|endofturn|]"
    output_texts.append(prompt)
  return output_texts
```

```python
def save_dataset_to_csv(dataset, file_path):
    """
    주어진 데이터셋을 CSV 파일로 저장합니다.

    매개변수:
    - dataset: 리스트의 리스트 형태로, 각 내부 리스트는 데이터의 한 행을 나타냅니다.
    - file_path: 저장할 CSV 파일의 경로와 이름을 포함한 문자열입니다.

    반환값:
    - 없음
    """
    # 데이터셋을 DataFrame으로 변환
    df = pd.DataFrame(dataset)

    # DataFrame을 CSV 파일로 저장
    df.to_csv(file_path, index=False, encoding='utf-8-sig')
```

```python
save_dataset_to_csv(traindata, "/content/data/careerboost_dataset.csv")
```
